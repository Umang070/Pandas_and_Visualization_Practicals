```python
drinks =  pd.read_csv("http://bit.ly/drinksbycountry")
drinks.dtypes
```

```python
#multiple col datatypes changes and astype in dataframe
drinks = drinks.astype({'beer_servings':'float','spirit_servings':'float'}).copy()
drinks.dtypes
```

```python
drinks=drinks.set_index('country')
drinks.head()
```
