## Dist Plot

```python
val = np.random.rand(100)
# val
```

```python
sns.distplot(val,color='red')  #for randn
```

```python
label_dist = pd.Series(val,name='Variable_x')
sns.distplot(label_dist, vertical=True, hist=False)
```

```python
# kdeplot : Show a univariate or bivariate distribution with a kernel
#           density estimate.
# rugplot : Draw small vertical lines to show each observation in a
#           distributio
sns.distplot(label_dist, rug=True, hist=True, rug_kws={"color": "r"},
                       kde_kws={"color": "k", "lw": 3, "label": "KDE"},
                       hist_kws={"histtype": "step", "linewidth": 3,
                                 "alpha": 0.7, "color": "b", "label": "HIST"})
```
