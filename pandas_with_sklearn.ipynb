```python
df = pd.read_csv('C:\\Users\\UMANG PATEL\\Desktop\\Scikit_learn\\diamonds.csv',index_col=0)
```

```python
# df.cut.astype('category').cat.codes
# df.cut.unique()
df.color.unique()
df.clarity.unique()
```

```python
cut_class = {'Fair': 1, 'Good': 2, 'Very Good': 3, 'Premium': 4, 'Ideal': 5}
clarity_class = {'I3': 1, 'I2': 2, 'I1': 3, 'SI2': 4, 'SI1': 5, 'VS2': 6, 'VS1': 7, 'VVS2': 8, 'VVS1': 9}
color_class ={'J': 1, 'I': 2, 'H': 3, 'G': 4, 'F': 5, 'E': 6, 'D': 7}
```

```python
df.cut = df.cut.map(cut_class)
df.clarity = df.clarity.map(clarity_class)
df.color = df.color.map(color_class)
```

```python
df
```

```python
import sklearn
from sklearn import svm, preprocessing
df.isnull().sum()
```

```python
df = sklearn.utils.shuffle(df)
df = df.dropna()
X = df.drop('price',axis=1).values
X = preprocessing.scale(X)
y = df.price.values

test_size = 200

X_train = X[:-test_size]
y_train = y[:-test_size]

X_test = X[-test_size:]
y_test = y[-test_size:]

clf = svm.SVR(kernel='linear')
clf.fit(X_train,y_train)
```
